## Function to parse the data
This function takes the line, splits it into person and friend list. After that it creates pairs of friends and for each pair append the friend list to the resultant list.

```
def json_parse_pair_list(line):
  person_list = line.split("\t")
  person = json.loads(person_list[0])
  friends = json.loads(person_list[1])
  result = []
  for i in range(len(friends)):
    f_list = [str(entry['id'])+":"+str(entry['name'])+":"+str(entry['hometown']) for entry in friends]
    del f_list[i]
    sorted_pair = sorted((person,friends[i]),key=lambda x:x['id'])
    pair = [str(entry['id'])+":"+str(entry['name'])+":"+str(entry['hometown']) for entry in sorted_pair]
    result.append((str(tuple(pair)),f_list))
  return result
```

## Applying the transformation

```
pair_list = lines.flatMap(json_parse_pair_list)
pair_list.take(1)
```

## Common Friends
This function takes the two lists and get the intersection of the list

```
def common_friends(flist1,flist2):
  if len(flist1) == 0 or len(flist2) == 0:
    return []
  result = []
  for outer in flist1:
    for inner in flist2:
      if outer == inner:
        result.append(outer)
  return result
```

## Applying the transformation
The following code applies the reduceByKey transformation on the rdd

```
pair_list.reduceByKey(common_friends)
pair_list.take(1)
```
